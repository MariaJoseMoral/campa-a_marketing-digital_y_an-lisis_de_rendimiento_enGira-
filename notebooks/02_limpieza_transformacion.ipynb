## 2. LIMPIEZA Y TRANSFORMACIÓN

- normalización 
- cambio del tipo dato (fecha)
- eliminar columnas 
- crear nuevas variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/raw/data.csv", sep=";", engine="python", skiprows=3)

df.head()
```

```python
df = df.drop(columns=[
    "Name from", "Email from", "Hard bounces", 
    "Soft bounces", "Non delivered", "Delivered", 
    "Unsubscribed", "Complaints"])
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
```

```python
df['sending_date'] = pd.to_datetime(df["sending_date"], dayfirst=True, errors="coerce")
```

```python
df['sending_date'].dtype
```

```python
df['year'] = df['sending_date'].dt.year
df['month'] = df['sending_date'].dt.month
df['day'] = df['sending_date'].dt.day
df['weekday'] = df['sending_date'].dt.day_name(locale='es_ES')
```

```python
df['subject'].value_counts()
```

```python
df['subject'] = df['subject'].str.lower().str.strip()
```

```python
def clasificar_campaña(nombre):
    
    if "convocatorias" in nombre:
        return "convocatorias"
    
    elif "participacion" in nombre or "participación" in nombre or "invitación" in nombre or "invitacio" in nombre:
        return "invitacion"
    
    elif "jornadas" in nombre:
        return "eventos"
    
    elif "demo" in nombre or 'webinar' in nombre or 'registro' in nombre:
        return "formacion"
    
    elif "dferia" in nombre or 'madferia' in nombre or 'mercartes' in nombre:
        return "ferias"
    
    elif "descubre" in nombre or "descobreix" in nombre or 'destacados' in nombre:
        return "promocion"
    
    elif "funcionamiento" in nombre or "resumen" in nombre or "fin" in nombre:
        return "informativa"
    
    else:
        return "otros"
```

```python
df['tipo_campaña'] = df['subject'].apply(clasificar_campaña)
```

```python
df['tipo_campaña'].value_counts()
```

```python
df['campaign_name'] = df['campaign_name'].str.lower().str.strip()
```

```python
def clasificar_publico(nombre):
    
    if 'artistas' in nombre or 'convocatorias' in nombre or 'webinar' in nombre or 'diagnostico'in nombre:
        
        return 'artistas'
    
    elif 'programadores' in nombre or 'demo' in nombre:
        
        return 'programadores'
    
    else:
        return 'all'
```

```python
df['tipo_publico'] = df['campaign_name'].apply(clasificar_publico)
```

```python
df['tipo_publico'].value_counts()
```

```python
df.columns
```

```python
df.sample(5)
```

```python
df.columns = df.columns.str.replace("-", "_")

cols_pct = [
    "non_delivered_rate",
    "trackable_open_rate",
    "click_rate",
    "click_to_open_rate",
    "unsubscription_rate",
    "delivered_rate",
    "hard_bounces_rate",
    "soft_bounces_rate",
    "complaints_rate"
]

for col in cols_pct:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("%", "")
        .str.replace(",", ".")
        .astype(float)
)
```

```python
df.to_csv("data_limpia.csv", index=False)
```
