## 📊 EDA – Newsletter performance (código completo)

```python
# ==============================
# 1. IMPORTS
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ==============================
# 2. CARGA DE DATOS
# ==============================
df = pd.read_csv("data_limpia.csv")

pd.set_option("display.max_columns", None)

# ==============================
# 3. OVERVIEW GENERAL
# ==============================

print("Shape:", df.shape)
display(df.head())
print(df.info())

# ==============================
# 4. NULOS
# ==============================
print(df.isnull().sum() / len(df) * 100)

# ==============================
# 5. DUPLICADOS
# ==============================
print("Duplicados:", df.duplicated().sum())

# ==============================
# 6. PRINCIPALES ESTADÍSTICOS
# ==============================
print(df.describe())


# ==============================
# 7. KPIs GENERALES
# ==============================

print("Open rate medio:", df["trackable_open_rate"].mean())
print("CTR medio:", df["click_rate"].mean())

# ==============================
# 8. ANÁLISIS POR TIPO DE CAMPAÑA
# ==============================

kpi_tipo = df.groupby("tipo_campaña")[["trackable_open_rate", "click_rate"]].mean().sort_values("trackable_open_rate", ascending=False)
display(kpi_tipo)

# ==============================
# 9. ANÁLISIS POR PÚBLICO
# ==============================

kpi_publico = df.groupby("tipo_publico")[["trackable_open_rate", "click_rate"]].mean()
display(kpi_publico)

# ==============================
# 10. ANÁLISIS CRUZADO
# ==============================

kpi_cross = df.groupby(["tipo_campaña", "tipo_publico"])[["trackable_open_rate", "click_rate"]].mean()
display(kpi_cross)

# ==============================
# 11. TOP CAMPAÑAS
# ==============================

top_open = df.sort_values("trackable_open_rate", ascending=False)[["campaign_name", "trackable_open_rate"]].head(10)
display(top_open)

top_ctr = df.sort_values("click_rate", ascending=False)[["campaign_name", "click_rate"]].head(10)
display(top_ctr)

# ==============================
# 12. VISUALIZACIONES
# ==============================

# Open rate por tipo
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="tipo_campaña", y="trackable_open_rate", estimator=np.mean)
plt.xticks(rotation=45)
plt.title("Open Rate medio por tipo de campaña")
plt.show()

# CTR por tipo
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="tipo_campaña", y="click_rate", estimator=np.mean)
plt.xticks(rotation=45)
plt.title("CTR medio por tipo de campaña")
plt.show()

# Heatmap tipo vs público
pivot = df.pivot_table(values="trackable_open_rate", index="tipo_campaña", columns="tipo_publico", aggfunc="mean")

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Open Rate por tipo de campaña y público")
plt.show()

# ==============================
# 12. INSIGHTS RÁPIDOS
# ==============================

print("\n--- INSIGHTS ---")
print("Mejor tipo campaña (open rate):")
print(kpi_tipo.index[0])

print("\nPeor tipo campaña:")
print(kpi_tipo.index[-1])
```
